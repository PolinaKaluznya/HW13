***Задача 3***

1. Выполните POST-запрос к к ресурсу `lookup`, чтобы найти первые 100 записей, содержащих альбомы для артиста с `amgArtistId`, равного  полученному ранее `amg_artist_id`. Укажите время ожидания запроса 60 секунд и распечатайте ответ в формате json.


```python
import requests
import json

# Замените "полученный ранее amg_artist_id" на фактическое значение
amg_artist_id = "полученный ранее amg_artist_id"

url = "https://lookup"  # или "http://lookup" если используется HTTP

payload = {
    "artistId": amg_artist_id,
    "limit": 100,
    "type": "albums"
}

headers = {
    'Content-Type': 'application/json'
}

response = requests.post(url, headers=headers, data=json.dumps(payload), timeout=60)

if response.status_code == 200:
    data = response.json()
    album_count = len(data["albums"])
    print(f"Количество записей: {album_count}")
else:
    print(f"Ошибка: {response.status_code} - {response.text}")
```
